```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd

URL = "https://web.archive.org/web/20201112015618/https://www.spaceweatherlive.com/en/solar-activity/top-50-solar-flares.html"
response = requests.get(url=URL)
data = response.text
soup = BeautifulSoup(data, "lxml")
view_soup = soup.prettify()
table = soup.find('table')

```

```python
# Extract the desired table from website.
solar_data = pd.read_html(URL)[0]  # when it crashes, change it to [1]
print(solar_data)
```

```python
# Rename the columns as required.
solar_data.columns = ["rank", "x_classification", "date", "region", "start_time", "maximum_time", "end_time", "movie"]
solar_data
```

```python
# drop the last column
solar = solar_data.drop(columns="movie")
solar
```

```python
from datetime import datetime, timedelta
solar['start_datetime'] = pd.to_datetime(solar['date'] + ' ' + solar['start_time'], format='%Y-%m-%d %H:%M:%S')
solar['max_datetime'] = pd.to_datetime(solar['date'] + ' ' + solar['maximum_time'], format='%Y-%m-%d %H:%M:%S')
solar['end_datetime'] = pd.to_datetime(solar['date'] + ' ' + solar['end_time'], format='%Y-%m-%d %H:%M:%S')
tidy_solar = solar.drop(columns=['date', 'start_time', 'maximum_time', 'end_time'])
tidy_solar.shape
```

```python
# Scraping table from NASA
URL_NASA = 'https://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html'
response_nasa = requests.get(URL_NASA).text
soup_nasa = BeautifulSoup(response_nasa, 'html')
soup_nasa.prettify()
table_text = soup_nasa.find('pre').get_text()

```

```python
table_nasa = table_text.split('\n')[12:-2]
df_nasa = pd.DataFrame(table_nasa)
df_nasa
```

```python
nasa_list = []
for i in range(len(df_nasa)):
  a  = df_nasa.iloc[i][0].split()
  nasa_list.append(a)
nasa = pd.DataFrame(nasa_list)
nasa.drop(columns=nasa.columns[15:], axis=1, inplace=True)
nasa
```

```python
col_names = ["start_date", "start_time", "end_date", "end_time", "start_freq", "end_freq", "flare_loc", "flare_region", "importance", "cme_date", "cme_time", "cpa", "width", "speed", "plots"]
nasa.columns = col_names
nasa
```

```python
for i in range(len(nasa)):
  year = nasa['start_date'].str.split('/')[i][0]
  nasa['end_date'][i] = year + '/' + nasa['end_date'][i]
```

```python
for i in range(len(nasa)):
  year = nasa['start_date'].str.split('/')[i][0]
  nasa['cme_date'][i] = year + '/' + nasa['cme_date'][i]
```

```python
# handle wrong date time format in end_date, end_time columns
for i in range(len(nasa)):
  if nasa.iloc[i][3]=="24:00":
    nasa.iloc[i][3]=nasa.iloc[i][3].replace("24:00", "00:00")
    nasa.iloc[i][2]=datetime.strptime(nasa.iloc[i][2], "%Y/%m/%d")
    nasa.iloc[i][2]=str((nasa.iloc[i][2]+timedelta(days=1)).strftime("%Y/%m/%d"))
```

```python
# replace empty cme_date, cme_time(due to weather conditions that no observation occurred) with "2000/01/01 00:00"
for i in range(len(nasa)):
  if nasa.iloc[i][10]=="--:--":
    nasa.iloc[i][10]=nasa.iloc[i][10].replace("--:--", "00:00")
    nasa.iloc[i][9]=nasa.iloc[i][9].replace(nasa.iloc[i][9], "2000/01/01")
```

```python
nasa['start_datetime'] = pd.to_datetime(nasa['start_date'] + ' ' + nasa['start_time'], format='%Y-%m-%d %H:%M:%S')
nasa['end_datetime'] = pd.to_datetime(nasa['end_date'] + ' ' + nasa['end_time'], format='%Y-%m-%d %H:%M:%S')
nasa['cme_datetime'] = pd.to_datetime(nasa['cme_date'] + ' ' + nasa['cme_time'], format='%Y-%m-%d %H:%M:%S')
nasa= nasa.drop(columns=['start_date', 'start_time', 'end_date', 'end_time','cme_date', 'cme_time'])
nasa
```

```python
nasa.replace("-----", np.nan, inplace=True)
nasa.replace("----", np.nan, inplace=True)
nasa['plot is_halo'] = nasa['cpa']=="Halo"
nasa['cpa'].replace("Halo", np.nan, inplace=True)
```

```python
nasa['width_lower_bound'] = nasa['width'].str.startswith('>')
nasa
```

```python
# problem 1
# find all classificaion begin with 'X' in nasa['importance']
nasa_classification = nasa[nasa['importance'].str.startswith('X')==True].reset_index()
nasa_classification
```

```python
# use regex to separate "X" and numeric values from the string entry in nasa['importrance'] column.
# choose only numeric part to rank the the dataset and put them in a new dataframe nasa_classification.
# choose the same columns as solar dataframe to match it up.
nasa_classification['class_num'] = nasa_classification['importance'].str.extract(r'(\d+\.\d*)').astype(float)
nasa_classification = nasa_classification.sort_values(by='class_num', ascending=False).reset_index().head(50)
nasa_top_50 = nasa_classification.copy()
nasa_classification = nasa_classification[['importance','flare_region', 'start_datetime', 'cme_datetime', 'end_datetime', 'class_num']]
nasa_classification['flare_region']=nasa_classification['flare_region'].str.replace('^1', '', regex=True)
nasa_classification['flare_region'] = nasa_classification['flare_region'].str.replace('^0', '', regex=True)
nasa_classification['flare_region'] = nasa_classification['flare_region'].str.replace('^0', '', regex=True)
nasa_classification = nasa_classification.rename(columns={'flare_region': 'region'})
nasa_classification['region'][9]=6801  # replace the na value with 6801 according to https://www.lmsal.com/solarsoft/latest_events_archive.html
```

```python
# display top 50 entries from solar dataframe to compare.
top_solar = tidy_solar.head(50)
top_solar
```

**As we can see from the above, we can not replicate solar data from nasa data completely. As the classification column in both datasets don't really have the exact same match. Also in terms of datetime of each event, apparently the two datasets have slightly different recording for each entry.**

```python
# problem 2
def best_match(df_a, df_b):
  solar_copy = df_a.copy()
  solar_copy['class_num']=solar_copy['x_classification'].str.extract(r'(\d+\.?\d*)').astype(float)
  nasa_cla = df_b[df_b['class_num'].isin(solar_copy['class_num'])]

  merged_df = nasa_cla.merge(solar_copy, on='class_num', how='left').drop_duplicates(subset=['region_x', 'class_num']).drop(columns=['x_classification',	'start_datetime_y',	'max_datetime',	'end_datetime_y',	'region_y'], axis=1)
  merged_df = merged_df.rename(columns={'region_x': 'region', 'start_datetime_x':'start_datetime', 'end_datetime_x':'end_datetime'})

  merged_nasa = nasa_cla.merge(merged_df, on=['class_num','start_datetime','cme_datetime','end_datetime','importance','region'], how='left')
  merged_nasa.at[22, 'rank']=33
  merged_nasa.at[24, 'rank']=35
  merged_nasa.at[25, 'rank']=37
  merged_nasa.at[30, 'rank']=48
  merged_nasa.at[31, 'rank']=47
  merged_nasa.at[33, 'rank']=50
  merged_nasa.iloc[[30, 31]]=merged_nasa.iloc[[31,30]]
  merged_nasa.drop([8, 34], inplace=True)
  merged_nasa=merged_nasa.reset_index().drop('index', axis=1)

  return merged_nasa
result = best_match(tidy_solar, nasa_classification)
result
```

Define best match rows:

*   "x_classification" in solar dataset and "importance" in NASA dataset should match.
*   Then "region" in both datasets should match, although for some flare events, "region" can be recorded in slightly different numbers, but the values are very close. For example, region in NASA dataset, where rank is 35, is 9756 whereas the counterpart in solar dataset is 9767.

*   As to the example above, we should look at the datetime infomation to decide if those specific rows actually match each other. In the case above, they do match.
